## Parameters and Data Import

```python
pi = 3.141592653589793238
```

```python
GelMA21 = pd.read_csv("GelMA21.csv")
GelMA21_H = 5.17
GelMA21_D = 18.55/1000
GelMA21_R = GelMA21_D/2

GelMA22 = pd.read_csv("GelMA22.csv")
GelMA22_H = 5.78
GelMA22_D = 18.52/1000
GelMA22_R = GelMA22_D/2

GelMA23 = pd.read_csv("GelMA23.csv")
GelMA23_H = 5.21
GelMA23_D = 18.18/1000
GelMA23_R = GelMA23_D/2

GelMA24 = pd.read_csv("GelMA24.csv")
GelMA24_H = 5.95
GelMA24_D = 18.44/1000
GelMA24_R = GelMA24_D/2

GelMA25 = pd.read_csv("GelMA25.csv")
GelMA25_H = 4.70
GelMA25_D = 17.95/1000
GelMA25_R = GelMA25_D/2
```

```python
GelMA21.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA22.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA23.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA24.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA25.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)
```

## Stress and Strain Calculation

```python
GelMA21['Engineering Strain'] = GelMA21['Stroke [mm]']/ GelMA21_H 
GelMA21['Stress [Pa]'] = GelMA21['Force [N]']/(pi*GelMA21_R**2)

GelMA22['Engineering Strain'] = GelMA22['Stroke [mm]']/ GelMA22_H 
GelMA22['Stress [Pa]'] = GelMA22['Force [N]']/(pi*GelMA22_R**2)

GelMA23['Engineering Strain'] = GelMA23['Stroke [mm]']/ GelMA23_H 
GelMA23['Stress [Pa]'] = GelMA23['Force [N]']/(pi*GelMA23_R**2)

GelMA24['Engineering Strain'] = GelMA24['Stroke [mm]']/ GelMA24_H 
GelMA24['Stress [Pa]'] = GelMA24['Force [N]']/(pi*GelMA24_R**2)

GelMA25['Engineering Strain'] = GelMA25['Stroke [mm]']/ GelMA25_H 
GelMA25['Stress [Pa]'] = GelMA25['Force [N]']/(pi*GelMA25_R**2)



GelMA21['Engineering Strain'] = GelMA21['Stroke [mm]']/ GelMA21_H 
GelMA21['Stress [kPa]'] = (GelMA21['Force [N]']/(pi*GelMA21_R**2))/1000

GelMA22['Engineering Strain'] = GelMA22['Stroke [mm]']/ GelMA22_H 
GelMA22['Stress [kPa]'] = (GelMA22['Force [N]']/(pi*GelMA22_R**2))/1000

GelMA23['Engineering Strain'] = GelMA23['Stroke [mm]']/ GelMA23_H 
GelMA23['Stress [kPa]'] = (GelMA23['Force [N]']/(pi*GelMA23_R**2))/1000

GelMA24['Engineering Strain'] = GelMA24['Stroke [mm]']/ GelMA24_H 
GelMA24['Stress [kPa]'] = (GelMA24['Force [N]']/(pi*GelMA24_R**2))/1000

GelMA25['Engineering Strain'] = GelMA25['Stroke [mm]']/ GelMA25_H 
GelMA25['Stress [kPa]'] = (GelMA25['Force [N]']/(pi*GelMA25_R**2))/1000
```

## Force_Stroke Margin Error

```python
def error_margin_fs (mean):
    a = 0.001/mean #1mm = 0.001m
    return (1 +- a)*mean
```
