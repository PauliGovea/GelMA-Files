```python
GelMA11.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA12.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA13.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA14.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)

GelMA15.rename(columns={"Force":"Force [N]",
                "Stroke":"Stroke [mm]", "Time":"Time [sec]"}
          ,inplace=True)
```

## Stress and Strain Calculation

```python
GelMA11['Engineering Strain'] = GelMA11['Stroke [mm]']/ GelMA11_H 
GelMA11['Stress [Pa]'] = GelMA11['Force [N]']/(pi*GelMA11_R**2)

GelMA12['Engineering Strain'] = GelMA12['Stroke [mm]']/ GelMA12_H 
GelMA12['Stress [Pa]'] = GelMA12['Force [N]']/(pi*GelMA12_R**2)

GelMA13['Engineering Strain'] = GelMA13['Stroke [mm]']/ GelMA13_H 
GelMA13['Stress [Pa]'] = GelMA13['Force [N]']/(pi*GelMA13_R**2)

GelMA14['Engineering Strain'] = GelMA14['Stroke [mm]']/ GelMA14_H 
GelMA14['Stress [Pa]'] = GelMA14['Force [N]']/(pi*GelMA14_R**2)

GelMA15['Engineering Strain'] = GelMA15['Stroke [mm]']/ GelMA15_H 
GelMA15['Stress [Pa]'] = GelMA15['Force [N]']/(pi*GelMA15_R**2)


GelMA11['Engineering Strain'] = GelMA11['Stroke [mm]']/ GelMA11_H 
GelMA11['Stress [kPa]'] = (GelMA11['Force [N]']/(pi*GelMA11_R**2))/1000

GelMA12['Engineering Strain'] = GelMA12['Stroke [mm]']/ GelMA12_H 
GelMA12['Stress [kPa]'] = (GelMA12['Force [N]']/(pi*GelMA12_R**2))/1000

GelMA13['Engineering Strain'] = GelMA13['Stroke [mm]']/ GelMA13_H 
GelMA13['Stress [kPa]'] = (GelMA13['Force [N]']/(pi*GelMA13_R**2))/1000

GelMA14['Engineering Strain'] = GelMA14['Stroke [mm]']/ GelMA14_H 
GelMA14['Stress [kPa]'] = (GelMA14['Force [N]']/(pi*GelMA14_R**2))/1000

GelMA15['Engineering Strain'] = GelMA15['Stroke [mm]']/ GelMA15_H 
GelMA15['Stress [kPa]'] = (GelMA15['Force [N]']/(pi*GelMA15_R**2))/1000
```
